```
% matplotlib inline
import matplotlib.pyplot as plt
import json
import nltk
import numpy as np
import pandas as pd
import seaborn as sns
from collections import Counter
```

```
data = pd.read_csv(root_dir + "/data/all_articles_dependencies.csv")
```

```
month_mapper = {
    7: "jul",
    8: "aug",
    10: "oct",
    11: "nov",
    12: "dec",
    1: "jan"
}
```

```
def get_month(date):
  return month_mapper[int(date.split("-")[1])]
```

```
data["month"] = data["publishedAt"].astype(str).apply(get_month)
```
